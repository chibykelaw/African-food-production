## What is the total number and percentage of missing data in 2016 to 2 decimal places?

## Which of these elements had the highest sum in 2018?

```python
#select columns 'Y2018' and 'Element'

df_y2018_element = df[['Y2018','Element']]
```

```python
df_y2018_element.groupby('Element').sum().sort_values(by='Y2018',ascending=False)
```

### Result:
    Domestic supply quantity (2161192.10) is the "Element" that had the highest sum 2018.

## Which of these elements had the 3rd lowest sum in 2018?

```python
df_y2018_element.groupby('Element').sum().sort_values(by='Y2018',ascending=True)
```

### Result:
    Protein supply quantity (g/capita/day) (11833.56) is the "Element" that had the 3rd lowest sum in 2018
